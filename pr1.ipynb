```python
#  dict
Rno = input("Enter the roll numbers for 3 students (using space): ").split()
Rno1 = list(map(int, Rno))
names = input("Enter the names of 3 students (using space): ").split()

students = dict(zip(Rno1, names))
print("Students with roll numbers:", students)
Rno2 = input("Enter the roll numbers for 2 students (using space): ").split()
Rno3 = list(map(int, Rno2))

names1 = input("Enter the names of 2 students (using space): ").split()

students1 = dict(zip(Rno3, names1))
print("Students with roll numbers:", students1)
students.update(students1)

print("Merged dictionary:", students)
```

```python
#functiion
num=int(input("Enter a number:"))

def factorial(num):
    return[i for i in range(1,num+1) if num%i==0]

print("The factorial of the number",num, " is:",factorial(num))
```

```python
# e using function return true if 2 list have atleast one common value
def common_member(list1, list2):
    for item in list1:
        if item in list2:
            return True
    return False

list1 = [4,7,8,2]
list2 = [3,4,2,5]

print(common_member(list1, list2))
```
